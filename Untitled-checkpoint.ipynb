```python
X= df.drop("quality", axis=1)
Y= df ['quality']
```

```python
seed=33
X_train, X_test, Y_train, Y_test= train_test_split(
X, Y, random_state=seed, test_size=0.2
)
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

```python
#VOILA mon jeu d'entrinement quand 5197 Observations et manger de test compte 1300 observtions
#maintenant il faut que je standardise mes données
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled
```

```python
#dans ce cas on a une variable dependante numérique donc on vaconstruire un modele de regression
#on va voire la moyenne quadratique de l'erreur pour choisir le meilleur modele
from sklearn.linear_model import LinearRegression

lr_cve = cross_val_score(
    LinearRegression(),
    X_train_scaled,
    Y_train,
    cv=3,
    scoring='neg_mean_squared_error'
)
from sklearn.ensemble import RandomForestRegressor

rf_cve = cross_val_score(
    RandomForestRegressor(random_state=seed),
    X_train_scaled,
    Y_train,
    cv=3,
    scoring='neg_mean_squared_error'
)
from sklearn.svm import SVR

svm_cve = cross_val_score(
    SVR(),
    X_train_scaled,
    Y_train,
    cv=3,
    scoring='neg_mean_squared_error'
)
```

```python
lr_cve
```

```python
rf_cve
```

```python
svm_cve
```

#lr c'est le modele le moins performant car il a 0.53; plus la valeur est grande plus la performance baisse

```python
print("RF: ", rf_cve.mean(), "LR: ", lr_cve.mean(), "SVM: ", svm_cve.mean())
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=seed)
model.fit(X_train_scaled, Y_train)
```

```python
Y_pred = model.predict(X_test_scaled)
Y_pred
```

```python
mean_squared_error(Y_test, Y_pred, squared=False)
#on a 0,63 donc ca veut dire que chaque score prédit par le modele le vrai score c'est plus ou moins 
#cette valeur la 0,63 ALORS ON PEUT DIRE QUE TEL VIN A UN SCORE DE 7 PAR EXEMPLE LE VRAIE RESULTAT SE SIUE ENTRE
#7+0,63 ET 7-0,63 ALORS C'EST ENTRE CES 2 VALEURS OU CE SITUE LA VRAIE VALEUR ET PLUS NOTRE ERREUR EST FAIBLE PLUS 
#NOTRE MODELE EST PERFORMANT
```
